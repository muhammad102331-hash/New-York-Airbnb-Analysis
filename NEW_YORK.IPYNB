```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# droping of missing values
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.dtypes
```

```python
df['id'] = df['id'].astype('object')
df.dtypes
```

```python
df['host_id'] = df['host_id'].astype('object')
df.dtypes
```

```python
# # idenfying outliers in price
df = df[df['price']< 1500]
sns.boxplot(data=df, x='price')
```

```python
# #Price distribuion
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='price', bins=100, kde=True)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.show
```

```python
df.dtypes
```

```python
# #Price distribuion
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='availability_365', bins=50, kde=True)
plt.title('availability_365 Distribution')
plt.xlabel('availability_365')
plt.show()
```

```python
df.groupby('neighbourhood_group')['price'].mean()
```

```python
df.dtypes
```

```python
df.columns
```

```python

df['price per bed']= df['price']/df['beds']
df.head()
```

```python
df.groupby(by='neighbourhood_group')['price per bed'].mean()
```

```python
# # price dependency on neighbourhood
sns.barplot(data=df, x='neighbourhood_group', y='price', hue='room_type')
```

```python
# number of reviews and price rel
plt.figure(figsize=(10,6))
plt.title('locality and reviews dependency')
sns.scatterplot(data=df, x='number_of_reviews', y='price', hue='neighbourhood_group')
plt.show()
```

```python
df.dtypes
```

```python
sns.pairplot(data=df, vars=['price', 'minimum_nights', 'number_of_reviews',  'availability_365'], hue='room_type')
```

```python
#Geographical Distribution of AirBnb Listing
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='longitude', y='latitude', hue='neighbourhood_group', alpha=0.5)
plt.title('Geographical Distribution of AirBnb Listing in New York City')
plt.show()
```

```python
## heat map - correlation of one variable with others for numerical column
corr = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',  'availability_365', 'beds', 'price per bed']].corr()
plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
